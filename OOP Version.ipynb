```python
import pandas as pd
import requests
import json
import requests
from oop_version import Trello
```

# Set up credentials

```python
credentials = pd.read_json('auth.json',typ='series')
```

```python
api_key = credentials['api_key']
api_token = credentials['token']
```

```python
board_id = 'xbbSYDmL'
```

```python
getTrello = Trello(api_key,api_token)
```

# 1. Board Description

```python
df_board = getTrello.getBoardDescription(board_id)
```

```python
df_board_info = pd.DataFrame([df_board['id'],df_board['name'],df_board['desc'],df_board['url']]).T
df_board_info.columns = ['board_id','name','desc','url']
df_board_info
```

# 2. List

```python
Lists=getTrello.getAllListOnABoard(board_id)
```

```python
AllLists = []
for k in range(len(Lists)):
    res = [Lists[0]['idBoard'],Lists[k]['id'],Lists[k]['name']]
    df_res = pd.DataFrame(res).T
    AllLists.append(df_res)
```

```python
df_AllLists = pd.concat(AllLists)
df_AllLists.columns = ['Board_id','List_id','List_Name']
df_AllLists
```

# 3. Cards

```python
list_id = list(df_AllLists['List_id'])
```

```python
AllCardsInsideList = [(k,getTrello.getAllCardsOnAList(k)) for k in list_id]
```

```python
AllCards =  []
for k in range(len(AllCardsInsideList)):
    listid = AllCardsInsideList[k][0]
    Cards = AllCardsInsideList[k][1]
    for j in range(len(Cards)):
        res = [listid,Cards[j]['id'],Cards[j]['desc'],Cards[j]['name'],Cards[j]['url']]
        df_res = pd.DataFrame(res).T
        AllCards.append(df_res)
```

```python
dfAllCards = pd.concat(AllCards)
dfAllCards.columns = ['List_id','Card_id','Card_Desc','Card_Name','Card_URL']
dfAllCards
```
